```python
import pandas as pd
import altair as alt

df = pd.read_excel("parkNums.xlsx")

click = alt.selection_multi(fields=["location"])


base = alt.Chart(df).encode(
    tooltip=['location'],
    theta=alt.Theta("parks", stack=True),
#     radius=alt.Radius("parks", scale=alt.Scale(type="sqrt", zero=True, rangeMin=20)),
    color="location",
     opacity=alt.condition(click, alt.value(1), alt.value(0.2)),

#     order=alt.Order(
#       # Sort the segments from least parks to most
#       'parks',
#       sort='ascending'
#     )
    
).properties(
    title = "Parks per Neighborhood"
)

c1 = base.mark_arc(innerRadius=20, stroke="#fff")

text = base.mark_text(radius=160, size=10).encode(text="parks")

(c1 + text).configure_range(
    category={'scheme': 'category20'}
).add_selection(click)
```

```python
df = pd.read_excel("parks3.xlsx")
df1 = pd.DataFrame(df).sort_values(by='Location', ascending=False)
feat_dropdown = alt.binding_select(options=[None,'Basketball','Bench', 'Dog Park', 'Handball', 'Multipurpose Courts','Netball','Ping Pong', 'Players Bench','Playground','Pump Track','Skate Park', 'Soccer', 'Spray Play', 'Street Hockey','Tennis'],labels=['All','Basketball','Bench', 'Dog Park', 'Handball', 'Multipurpose Courts','Netball','Ping Pong', 'Players Bench','Playground','Pump Track','Skate Park', 'Soccer', 'Spray Play', 'Street Hockey','Tennis'])

feat_select = alt.selection_single(
    fields=["Feature"],
    bind=feat_dropdown,
    name=" "
)

bars = alt.Chart(df1).mark_bar().encode(
    tooltip=["Feature","Park"],
    y='Location',
    x='count(Feature):Q',
#     y=alt.Y('Location:N', sort='ascending'),
    color='Feature',
#     order=alt.Order(
#       # Sort the segments from least parks to most
#       'count():Q',
#       sort='ascending'
#     )
).add_selection(feat_select
).transform_filter(
    feat_select)

text = alt.Chart(df1).mark_text(dx=-5.4, dy=3, color='white').encode(
    tooltip=["Feature"],
    x=alt.X('count():Q', stack='zero', title='Total Features'),
    y=alt.Y('Location'),
    detail='Feature:N',
    text=alt.Text('count():Q', format='.0f')
).properties(width=700).transform_filter(
    feat_select)

(bars + text).configure_range(
    category={'scheme': 'category20'}
).properties(
    title = "Park Features by Neighborhood")

```
